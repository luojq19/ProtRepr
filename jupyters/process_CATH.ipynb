```python
import pandas as pd

all_data = pd.read_csv('../data/swissprot/swissprot.tsv', sep='\t')
# select columns 'Entry' and 'Gene3D'
cath_data = all_data[['Entry', 'Gene3D']]
cath_data = cath_data[cath_data['Gene3D'].notnull()]
cath_data
```

```python
import re

def extract_cath_labels(s):
    pattern = r'\d+\.\d+\.\d+\.\d+'
    cath_list = re.findall(pattern, s)
    return ';'.join(cath_list)

cath_data['Gene3D'] = cath_data['Gene3D'].apply(extract_cath_labels)
display(cath_data)
# rename Gene3D to cath
cath_data = cath_data.rename(columns={'Gene3D': 'cath'})
print(len(cath_data))
cath_data.to_csv('../data/cath/swissprot_cath.csv', index=False)
```

```python
import pandas as pd
 
cath_data = pd.read_csv('../data/cath/swissprot_cath.csv')
display(cath_data)
# filter out single-label entries
cath_data['num_labels'] = cath_data['cath'].apply(lambda x: len(x.split(';')))
cath_data_single_label = cath_data[cath_data['num_labels'] == 1]
display(cath_data_single_label)
print(len(cath_data_single_label))
cath_data_single_label.to_csv('../data/cath/swissprot_cath_single_label.csv', index=False)
```

```python
import pandas as pd
import torch
from tqdm.auto import tqdm

cath_data = pd.read_csv('../data/cath/swissprot_cath_single_label.csv')
cath2occurence = {}
pids, labels = cath_data['Entry'].tolist(), cath_data['cath'].tolist()
for pid, label in zip(pids, labels):
    if label not in cath2occurence:
        cath2occurence[label] = 0
    cath2occurence[label] += 1
# filter out rows with more than 10 occurence
cath_data_filtered = cath_data[cath_data['cath'].apply(lambda x: cath2occurence[x] > 10)]

pid2embeddings = torch.load('../data/embeddings/sprot_10_1022_esm2_t33.pt')
print(len(pid2embeddings))
pid_with_emb = list(pid2embeddings.keys())
pid2cath = {pid: cath for pid, cath in zip(cath_data_filtered['Entry'].tolist(), cath_data_filtered['cath'].tolist())}
data = {}
for pid, cath in tqdm(pid2cath.items()):
    try:
        data[pid] = {'embedding': pid2embeddings[pid], 'cath': [cath]}
    except:
        continue
print(len(data))
torch.save(data, '../data/cath/sprot_10_1022_esm2_t33_cath_above_10_single_label.pt')
```

```python
# split train test val
import torch
import random

random.seed(42)
cath_data = torch.load('../data/cath/sprot_10_1022_esm2_t33_cath_above_10_single_label.pt')
pids = list(cath_data.keys())
random.shuffle(pids)
train_data = {pid: cath_data[pid] for pid in pids[:int(0.8 * len(pids))]}
val_data = {pid: cath_data[pid] for pid in pids[int(0.8 * len(pids)):int(0.9 * len(pids))]}
test_data = {pid: cath_data[pid] for pid in pids[int(0.9 * len(pids)):]}
print(len(train_data), len(val_data), len(test_data))
torch.save(train_data, '../data/cath/sprot_10_1022_esm2_t33_cath_above_10_single_label_train.pt')
torch.save(val_data, '../data/cath/sprot_10_1022_esm2_t33_cath_above_10_single_label_val.pt')
torch.save(test_data, '../data/cath/sprot_10_1022_esm2_t33_cath_above_10_single_label_test.pt')
```
